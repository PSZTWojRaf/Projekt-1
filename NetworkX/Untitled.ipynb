```python
import networkx as nx
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_nodes_from([2,3])
```

```python
nx.draw_networkx(G,node_color='green', node_size=300)
```

```python
G.add_edge(1,2)
```

```python
nx.draw_networkx(G)
```

```python
e = (2,3)
G.add_edge(*e)
```

```python
G.add_edges_from([(1,2),(1,3)])
```

```python
nx.draw_networkx(G)
```

```python
print(G.number_of_nodes())
```

```python
print(G.edges())
```

```python
list(G.neighbors(1))
```

```python
list(G.neighbors(1))
```

```python
print(G.degree())
```

```python
print(G.has_edge(1,2))
```

```python
print(G.has_node(2))
```
